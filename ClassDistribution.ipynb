```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

### (i) Height Distribution

```python
height_values = [5.4, 5.6, 5.8, 5.10, 6.0]
height_data = np.random.choice(height_values, 70) # 70 students in class
```

```python
height_data
```

```python
plt.hist(height_data, bins=5, edgecolor='black')
plt.title('Height Distribution')
plt.xlabel('Height (ft)')
plt.ylabel('Frequency')
plt.show()
```

### (ii) Weight Distribution

```python
weight_values = [50, 60, 65, 70, 75]
weight_data = np.random.choice(weight_values, 70)
```

```python
weight_data
```

```python
plt.hist(weight_data, bins=5, edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

### (iii) Joint Distribution

```python
joint_data = pd.DataFrame({'Height': height_data, 'Weight': weight_data})
```

```python
joint_data
```

```python
plt.scatter(joint_data['Height'], joint_data['Weight'])
plt.title('Joint Distribution of Height and Weight')
plt.xlabel('Height (ft)')
plt.ylabel('Weight (kg)')
plt.show()
```

### (iv) Statistical Parameters for Weight and Height data

```python
height_mean = np.mean(height_data)

height_median = np.median(height_data)

height_unique, height_counts = np.unique(height_data, return_counts=True)
height_mode = height_unique[np.argmax(height_counts)]

height_std = np.std(height_data)

height_var = np.var(height_data)

height_skew = pd.Series(height_data).skew()

height_kurtosis = pd.Series(height_data).kurtosis()


weight_mean = np.mean(weight_data)

weight_median = np.median(weight_data)

weight_unique, weight_counts = np.unique(weight_data, return_counts=True)
weight_mode = weight_unique[np.argmax(weight_counts)]

weight_std = np.std(weight_data)

weight_var = np.var(weight_data)

weight_skew = pd.Series(weight_data).skew()

weight_kurtosis = pd.Series(weight_data).kurtosis()
```

```python
print('Height -')
print(f'\tMean: {height_mean}')
print(f'\tMedian: {height_median}')
print(f'\tMode: {height_mode}')
print(f'\tStandard Deviation: {height_std}')
print(f'\tVariance: {height_var}')
print(f'\tSkewness: {height_skew}')
print(f'\tKurtosis: {height_kurtosis}\n')

print('Weight -')
print(f'\tMean: {weight_mean}')
print(f'\tMedian: {weight_median}')
print(f'\tMode: {weight_mode}')
print(f'\tStandard Deviation: {weight_std}')
print(f'\tVariance: {weight_var}')
print(f'\tSkewness: {weight_skew}')
print(f'\tKurtosis: {weight_kurtosis}')
```
